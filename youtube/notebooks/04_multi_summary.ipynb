# 멀티요약 비교

```python
import pandas as pd
from collections import Counter
import re
import matplotlib.pyplot as plt
```

```python
# 📥 1. 요약 결과 불러오기
summary_df = pd.read_csv("../data/summaries.csv")
summary_df.dropna(subset=["summary"], inplace=True)
```

```python
# 🔍 2. 각 영상 요약 간략 출력
for idx, row in summary_df.iterrows():
    print(f"\n🎬 {row['title']} ({row['category']})")
    print(f"📌 요약문:")
    print(row["summary"][:300] + "...")
    print("-" * 100)
```

# 키워드 분석 X

```python
# 🧠 3. 전체 요약 텍스트 합치기 → 키워드 분석용
all_text = " ".join(summary_df["summary"].tolist())
all_text = re.sub(r"[^\w가-힣\s]", "", all_text)  # 특수문자 제거
```

```python
# 📌 4. 조사 제거 및 불용어 필터링 (간단한 버전)
stopwords = ["입니다", "있습니다", "하는", "그리고", "하지만", "또한", "그는", "그녀는", "것은", "에서는", "으로", "이다", "은", "는", "이가", "의"]
words = [w for w in all_text.split() if w not in stopwords and len(w) > 1]
```

```python
# 📊 5. 빈도수 분석 및 시각화
common_words = Counter(words).most_common(20)
```

```python
print("\n📈 상위 키워드:")
for word, count in common_words:
    print(f"{word}: {count}회")
```
