```python
# ğŸ§  3. ì „ì²´ ìš”ì•½ í…ìŠ¤íŠ¸ í•©ì¹˜ê¸° â†’ í‚¤ì›Œë“œ ë¶„ì„ìš©
all_text = " ".join(summary_df["summary"].tolist())
all_text = re.sub(r"[^\wê°€-í£\s]", "", all_text)  # íŠ¹ìˆ˜ë¬¸ì ì œê±°
```

```python
# ğŸ“Œ 4. ì¡°ì‚¬ ì œê±° ë° ë¶ˆìš©ì–´ í•„í„°ë§ (ê°„ë‹¨í•œ ë²„ì „)
stopwords = ["ì…ë‹ˆë‹¤", "ìˆìŠµë‹ˆë‹¤", "í•˜ëŠ”", "ê·¸ë¦¬ê³ ", "í•˜ì§€ë§Œ", "ë˜í•œ", "ê·¸ëŠ”", "ê·¸ë…€ëŠ”", "ê²ƒì€", "ì—ì„œëŠ”", "ìœ¼ë¡œ", "ì´ë‹¤", "ì€", "ëŠ”", "ì´ê°€", "ì˜"]
words = [w for w in all_text.split() if w not in stopwords and len(w) > 1]
```

```python
# ğŸ“Š 5. ë¹ˆë„ìˆ˜ ë¶„ì„ ë° ì‹œê°í™”
common_words = Counter(words).most_common(20)
```

```python
print("\nğŸ“ˆ ìƒìœ„ í‚¤ì›Œë“œ:")
for word, count in common_words:
    print(f"{word}: {count}íšŒ")
```
